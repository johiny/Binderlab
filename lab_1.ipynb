```python
import altair as alt
import pandas as pd 
from vega_datasets import data as vega_data
data = vega_data.gapminder()
```

## Data

```python
lowest_lx_countries = data[data['country'].isin(['Rwanda', 'Afghanistan'])]
```

```python
lowest_lx_countries
```

```python
plot1 = alt.Chart(lowest_lx_countries).mark_line().encode(
    x='year',
    y='life_expect',
    color='country'
)
```

```python
global_lx_mean_by_year = pd.DataFrame(data.groupby('year')['life_expect'].mean()).reset_index()
global_lx_mean_by_year
```

```python
plot2 = alt.Chart(global_lx_mean_by_year).mark_line(size=10).transform_fold(
    fold=['Global Year Mean'],
).encode(
    x='year',
    y='life_expect',
    color='key:O'
)
plot2
```

```python
last_plot = plot1 + plot2
last_plot.encode(
alt.Color(legend=alt.Legend(title="Country And Global Mean"))
)
```
